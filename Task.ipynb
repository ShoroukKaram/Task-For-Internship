```python
from ntscraper import Nitter
```

```python
scraper = Nitter(0)
```

```python
import pandas as pd
```

```python
def count_mentions(user_name,stock_symbol):
    tweets = scraper.get_tweets(user_name, mode = 'user', number=100)
    final_tweets = []
    for x in tweets['tweets']:
        data = [ x['text']]
        final_tweets.append(data)
        dat = pd.DataFrame(final_tweets, columns =['text'])
        dat['mentions'] = dat['text'].apply(lambda text: text.count(stock_symbol))
        total_mentions=dat['mentions'].sum()
    return total_mentions

    
```

```python
print(count_mentions("Mr_Derivatives","$SOFI"))
```
